```python
import pandas as pd
import nltk
import numpy as np
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
filepath = 'Telemachus.txt'
text = ''
with open(filepath, 'r') as f:
  text = f.read()
```

```python
cachedStopWords = stopwords.words("english")
text = ' '.join([word for word in text.split() if word not in cachedStopWords])
text = ''.join([i for i in text if not i.isdigit()])
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
pdText = pd.Series(text.split())
pdText
```

```python
pdText = pdText.str.replace('[^\w\s]','').replace("in'",'ing')
```

```python
pdText = pdText[pdText.apply(lambda word: not word.isdigit())]
```

```python
pdText
```

```python
nltk.download('wordnet')#downloads dataset
lemma = WordNetLemmatizer()
unq_words = []
uw = {lemma.lemmatize(word) for word in pdText}
unq_words.append(uw)
y2 = [len(x) for x in unq_words]# to s
```
